### Создание заданных массивов
В данном разделе будут изучены подходы создания массив numpy с уже известным содержимым этих массивов.

Например, мы хотим создать одномерный массив с элементами:1,2,3. Используя стандартные средства python это будет выглядеть следующим образом:

```python
my_list = [1,2,3]
print(f'Инициализируя наш массив {my_list} таким образом, он будет иметь тип данных: {type(my_list)}')
```

list - это стандартный для python тип данных. Однако данный тип часто неудобен в обращении. Чтобы воспользоваться большинством удобных функций библиотеки numpy, имеющийся тип данных "list" нужно трансформировать в "numpy.ndarray" тип данных. 

```python
np_array = np.array(object=my_list)
print(f'Теперь наш массив {np_array} имеет тип {type(np_array)}')
```

На самом деле, мы могли сразу инициализировать массив с типом данных numpy.ndarray:

```python
array = np.array(1,2,3)
print(type(array))
print(array)
```

Но появилась ошибка, почему? Ответ: когда мы вызываем функцию np.array() она на входит принимает уже сам массив, а не его отдельные элементы, то есть:

```python
array = np.array(object=[1,2,3]) # TODO array = np.array(...)
print(type(array))
print(array)
```

### Работа с np.nan, inf и другими константами

np.nan -- по стандарту IEEE 754  представление с плавующей точкой отсутсвия значения. Или по простому, значение: "нет числа"

```python
np.nan
```

Некоторые особенности:

```python
print('4==4:',4==4)
print('np.nan == np.nan:',np.nan == np.nan)
print('np.isnan(np.nan):',np.isnan(np.nan))
print('np.isnan(np.inf):',np.isnan(np.inf))
```

В реальных задачах со структурированными данными, часто бывают пропущены или повреждены значения в массивах, которые и становятся np.nan. Это мешает работе некоторым алгоритмам, поэтому важно уметь с ними работать. Например, можно все nanы заменить на определенное число:

```python
array = np.array([1,2,np.nan,4,np.nan,np.inf])
array[np.isnan(array)] = 7
array
```

Так как inf, это другое значение элемента, то его тоже нужно учесть:

```python
array = np.array([1,2,np.nan,4,np.nan,np.inf])
array[np.isnan(array) | np.isinf(array)] = 7
array
```

Иногда, необходимо найти все нулевые значения или, наоборот, ненулевые: 

```python
indices = np.nonzero([1,2,0,0,4,0])
print(indices)
```

# Решение задачи линейной регресии

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def linear(x,k,b):
    return k*x+b
```

В нашей задаче будет всего один признак: x

```python
x = np.linspace(-3,12,101)
x
```

Создадим переменную y_noise, которая будет содержать в себе основную составляющую -- линейную регресиию с k=2 и b=5 и шум, который всегда присутсвует в реальных задачах.  

```python
y = linear(x,k=2,b=5)

sigma=6
mu = 0
noise = sigma * np.random.standard_normal(y.shape) + mu

y_noise= y + noise
```

В итоге получим следующую картину: 

```python
plt.plot(x,y,label='signal')
plt.grid()
plt.plot(x,y_noise,'o',markersize=3,label='noise signal')
plt.axvline(0,c='black')
plt.axhline(0,c='black')
plt.legend()
plt.show()
```

Линейная регрессия, это модель, результат которой линейно зависит от признаков и специального смещения. 
Модель линейной регрессии в матричном виде выглядит следующим образом:
$$
\mathbf{Y}' = \mathbf{X} \mathbf{W}
$$
где $\mathbf{Y}'$ - это вектор прогнозирумых значений, $\mathbf{X}$ - матрица исходных значений всех признаков. То есть, всего d признаков, в каждом по m значений, $\mathbf{W}$ - вектор значений весов.
$$
\mathbf{X}=\left[\begin{array}{ll}
\mathbf{x}_{1} & 1 \\
\vdots & \vdots \\
\mathbf{x}_{m} & 1
\end{array}\right] \in \mathbb{R}^{m \times(d+1)}, \mathbf{W}=\left[\begin{array}{l}
w_{1} \\
\vdots \\
w_{d} \\
b
\end{array}\right], \mathbf{Y}=\left[\begin{array}{l}
y_{1} \\
\vdots \\
y_{m}
\end{array}\right]
$$

Для того чтобы корректно пронгозировать $\mathbf{Y}'$ нужно иметь корректный вектор весов. Для это нужно решить задачу минимизации ошибки прогноза, зависящей только от этого вектора весов:

$$
F(\mathbf{W})=\frac{1}{m}\|\mathbf{Y}'-\mathbf{Y}\|^{2}=\frac{1}{m}\|\mathbf{X} \mathbf{W}-\mathbf{Y}\|^{2}
$$
 
Нам повезло, и решение данной задачи (методом наименьших квадартов) есть в матричном виде:
$$
\mathbf{W}=\left(\mathbf{X}^{\mathrm{T}} \mathbf{X}\right)^{-1} \mathbf{X}^{\mathrm{T}} \mathbf{Y}
$$

Теперь реализуем решение данной задачи в коде (у нас есть зашумленные данные, нам нужно найти аналитическую составляющую, которая в будщем позволит нам корректно прогнозировать примеры). 

Нам нужно в первую очередь создать матрицу X, у которого последний стоблец равен всегда 1. Вопрос*: почему? 

```python
x_1column = x.reshape(-1,1)
x_2column = np.ones(x_1column.shape[0]).reshape(-1,1) # как можно по другому
X = np.concatenate([x_1column,x_2column],axis=1)
print(X.shape)
print(X[:5])
```

Теперь создадим матрицу Y

```python
Y = y_noise.reshape(-1,1)
```

Теперь найдем корректные веса, из формулы, которая была дана выше:

```python
W = (np.linalg.inv(X.T @ X)) @ X.T @ Y
```

```python
W
```

Спрогнозируем велечины на исходных данных X:

```python
Y_predicted = X @ W
y_predicted = Y_predicted.ravel()
```

```python
plt.plot(x,y,label='signal')
plt.grid()
plt.plot(x,y_noise,'o',markersize=3,label='noise signal')
plt.plot(x,y_predicted,label='forecast')
plt.axvline(0,c='black')
plt.axhline(0,c='black')
plt.legend()
plt.show()
```

Как мы видим, мы успешно извлекли ("forecast") аналитическую составляющую "signal" из зашумленных данных "noise signal" при этом не имея вообще информации о значениях "signal".
