# Melt / Pivot

```python
df_melt = pd.melt(df.reset_index(), id_vars=['Параметр', 'РЕЖИМ'])
df_melt
```

```python
# df_melt[['РЕЖИМ','Параметр']].groupby('Параметр').mean()
```

```python
pd.pivot_table(data=df_melt,
    values=['value'],
    index=['Параметр','РЕЖИМ'],
    columns=['variable'],
    aggfunc='mean').reset_index(level=1)
```
