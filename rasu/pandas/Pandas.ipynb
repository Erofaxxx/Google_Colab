# Pandas (пример использования библиотеки на датасете Titanic)

## Pandas
Pandas – библиотека python, для работы с табличными данными.
Библиотека Pandas предоставляет широкие возможности для первичного анализа и обработки табличных данных и является одним из инструментов в арсенале Data Science специалиста. 


<img src='docs/img/panda start.jpg' width="800" height="600" >
<br>

__К основным функциям библиотеки можно отнести:__

*	Чтение данных из разных источников, создание табличных данных, сохранение данных
*	Анализ данных: отображение данных, расчет различных статистик, группировка данных, визуализация данных, …
*	Обработка данных: применение различных преобразований значений в таблице, сортировка, фильтрация, заполнение пропусков, расчет агрегированных параметров, создание подвыборок, объединение датасетов, …

__В Pandas реализовано две структуры хранения данных, с которыми вам предстоит работать:__
*	Series – это проиндексированный одномерный массив значений. Он похож на простой словарь типа dict, где имя элемента будет соответствовать индексу, а значение – значению записи.
*	DataFrame — это проиндексированный многомерный массив значений, соответственно каждый столбец DataFrame, является структурой Series.

В качестве индексов таблицы используются значения, начинающиеся с 0 и увеличивающиеся на 1 в каждой последующей строке. Также в данных содержится столбец PassengerId, который можно использовать в качестве индекса.

Мы можем выбрать значения с определенными именами индексов и колонок используя __метод loc()__

# Анализ данных

Библиотека Pandas содержит в себе большое количество методов для анализа данных.

Мы рассмотрим некоторые из них

С помощью __метода describe()__ можно получить описательные статистики для каждой из колонок. Если указать аргумент include='all' то будут выведены статистики как для числовых, так и для текстовых колонок (по умолчанию считается только для числовых)

```python
data.describe(include='all')
```

Количество уникальных значений в колонках - используем __метод nunique()__

```python
data.nunique()
```

Если нас интересует какие пассажирские классы присутствуют в датасете используем следующую конструкцию

```python
data['Pclass'].unique()
```

__value_counts()__ показывает как часто встречается каждое из значений в конкретной колонке.
Например, чтобы посмотреть сколько пассажиров находилось в каждом пассажирском классе, используем следующий код

```python
data['Pclass'].value_counts()
```

Pandas позволяет считать различные агрегации по столбцам, строкам и по всему датасету.

Наиболее часто используемые из них:
* сумма
* среднее значение 
* медиана
* перцентили

Подсчет суммарной стоимости билетов

```python
data['Fare'].sum()
```

Среднее значение стоимости билета

```python
data['Fare'].mean()
```

Медиана стоимости билета

```python
data['Fare'].median()
```

0.05 и 0.95 перцентили стоимости билетов

```python
data['Fare'].quantile(q=[0.05, 0.95])
```

Используя __функцию pd.crosstab__, можно посмотреть на пересечение значений в разных колонках.

Например, сколько пассажиров из каждого класса выжило

```python
pd.crosstab(data['Survived'], data['Pclass'])
```

Если разделить значения в предыдущей таблице на количество человек в каждом классе и умножить на 100, то можно посчитать процент выживших в каждом классе

```python
pd.crosstab(data['Survived'], data['Pclass']) / pd.crosstab(data['Survived'], data['Pclass']).sum() * 100
```

Аналогичным способом можно посчитать процент выживших в зависимости от пола

```python
pd.crosstab(data['Survived'], data['Sex']) / pd.crosstab(data['Survived'], data['Sex']).sum() * 100
```

Использование аргументов normalize=True и margins=True позволяет посчитать долю выживших пассажиров в каждой из категорий класса, признака Survived и их пересечений. 

При умножении на 100 мы получаем вместо доли процент

```python
pd.crosstab(data['Survived'], data['Pclass'], normalize=True, margins=True) * 100
```

# Обработка данных

Библиотека Pandas позволяет обрабатывать данные различными методами.

Абсолютные значения в колонках - __метод abs()__

```python
data['Fare'].abs()
```

Округление значений - __метод round()__

```python
data['Fare'].round()
```

Изменение типа (при переводи из float в int происходит округление в меньшую сторону) - __метод astype()__

```python
data['Fare'].astype(int)
```

__Метод diff()__ - разница текущего и предыдущего значения в колонках.
Часто бывает полезна при работе с временными рядами, например, резкие изменения какого-либо параметра могут говорить об 
определенных проблемах с оборудованием

```python
data['Fare'].diff()
```

__Метод cumsum()__ - сумма нарастающим итогом (бывает полезна при работе с временными рядами, например подсчет времени наработки узлов)

```python
data['Fare'].cumsum()
```

__Метод apply()__ - используется для применения функции к каждому элементу колонки

В данном примере мы логарифмируем все элементы из колонки Fare

```python
data['Fare'].apply(lambda x: np.log(x))
```

Мы можем создавать новые колонки используя различные комбинации существующих колонок.

Например, здесь мы создаем новый признак "Magic Feature" вычитая из стоимости билета возраст пассажира (существует не нулевая вероятность, что такой признак будет плохо работать при использовании в моделях)

```python
data['Magic Feature'] = data['Fare'] - data['Age']
data.head(3)
```

# Работа с пропусками

Довольно часто в реальных задача встречается ситуация, когда мы имеем пропуски в данных (значения в колонках отсутствуют).

Для выявления и обработки таких случаев в pandas имеется несколько методов:
* __isna()__ - проверяет, является ла значение пропуском
* __dropna()__ - удаляет строки или столбцы имеющие пропуски
* __fillna()__ - заполняет пропуски значениями (поддерживает различные методы заполнения)
* __interpolate()__ - интерполирует пропущенные значения

Подсчет количества пропущенных значений в каждой колонке

```python
data.isna().sum()
```

Удаление строк с пропущенными данными в колонке с возрастом

```python
data.dropna(subset=['Age'])
```

Удаление всех строк где пропущено значение в любой из колонок

```python
data.dropna()
```

Заполнение всех пропусков определенным значением

```python
data.fillna(-777)
```

Заполнение пропусков предыдущими значениями в колонке

```python
data.fillna(method='ffill')
```

# Группировка данных

__Метод groupby()__ позволяет группировать данные по определенным значениям в колонках и производить с этими группами различные манипуляции, например считать статистику по группам

Средние значения признаков в каждом пассажирском классе

```python
data.groupby('Pclass').mean()
```

Группировку можно производить по нескольким признакам.

В данном примере мы группируем по пассажирскому классу и полу и рассчитываем средние значения признаков 

```python
data.groupby(['Pclass', 'Sex']).mean()
```
